```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from fanalysis.ca import CA
```

```python
ar = np.array([[1.1, 2, 3.3, 4], [2.7, 10, 5.4, 7], [5.3, 9, 1.5, 15]])
df = pd.DataFrame(ar, index = ['a1', 'a2', 'a3'], columns = ['A', 'B', 'C', 'D'])
```

```python
values = np.array([[119, 26, 7], [54, 14, 10], [29, 14, 16],[84,17,64]])
df = pd.DataFrame(values, index = ['Marrons', 'Noisette', 'Verts','Bleus'], columns = ['Chatains', 'Roux', 'Blonds'])
                      
```
